```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AED - dados.csv')
display(df)
```

```python
df.groupby('departamento')['departamento'].count()
```

```python
df.groupby('departamento')['departamento'].count().rename('Quantidade de Funcionários')
```

```python
df.groupby('departamento').agg(qtde=('departamento', 'count')).sort_values('qtde', ascending=False).head(10)
```

```python
plt.bar(df['departamento'], df['salario'])
plt.show()
```

```python
plt.bar(df['departamento'], df['salario'])
plt.title('Salário por departamento')
plt.xlabel('Departamentos')
plt.ylabel('Salários')
plt.show()
```

```python
plt.pie(df['cargo'].value_counts(), labels=df['cargo'].unique())
plt.title('Distribuição de cargos')
plt.show()
```

```python
plt.pie(df['cargo'].value_counts(), labels=df['cargo'].unique(), autopct='%1.1f%%')
plt.title('Distribuição de cargos')
plt.show()
```

```python
explode = [0,0,0.1,0]
plt.pie(df['cargo'].value_counts(), labels=df['cargo'].unique(), autopct='%1.1f%%', explode=explode)
plt.title('Distribuição de cargos')
plt.show()
```
