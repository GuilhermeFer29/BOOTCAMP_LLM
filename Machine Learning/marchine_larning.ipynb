### TIPOS DE APREDIZADO DE MAQUINA SUPERVISIONADO e N√ÉO SUPERVISIONADO

```python
"""
O que √© aprendizado de m√°quina?
Imagine que voc√™ est√° ensinando uma crian√ßa a reconhecer frutas.
Voc√™ mostra v√°rias ma√ß√£s e diz: ‚ÄúIsso √© uma ma√ß√£.‚Äù Depois mostra bananas,
laranjas, etc. Com o tempo, a crian√ßa aprende a reconhecer sozinha o que √©
uma ma√ß√£, mesmo que nunca tenha visto aquela ma√ß√£ espec√≠fica antes.
O aprendizado de m√°quina (ou machine learning) funciona de forma parecida ‚Äî
√© como ensinar um computador a aprender com exemplos em vez de programar tudo
passo a passo. O computador aprende com os dados que voc√™ fornece e, com o tempo,
consegue prever o que acontece. Por exemplo, ele pode aprender a reconhecer
um e-mail de spam, a prever o tempo, a identificar pessoas em fotos, a jogar
xadrez, a dirigir um carro, a traduzir textos, a compor m√∫sicas, a fazer diagn√≥sticos
m√©dicos, a recomendar filmes, a vencer jogos de tabuleiro, a criar arte, a escrever
textos, a criar novos materiais, a descobrir novos medicamentos, a otimizar
processos industriais, a detectar fraudes, a fazer trading, a personalizar
um site, etc.


Exemplo simples:
Suponha que voc√™ queira que o computador descubra se uma fruta √© uma ma√ß√£ ou
uma banana. Em vez de escrever regras como ‚Äúse for vermelha e redonda, √© ma√ß√£‚Äù,
voc√™ mostra v√°rias frutas com suas caracter√≠sticas (cor, tamanho, formato) e o
nome da fruta.
O computador ent√£o analisa os dados e aprende padr√µes, como:
Frutas vermelhas e redondas costumam ser ma√ß√£s,


Frutas amarelas e compridas s√£o, geralmente, bananas.


Depois disso, voc√™ pode mostrar uma nova fruta e ele vai tentar adivinhar o que
√© com base no que aprendeu antes.



"""
```

### Regress√£o Linear

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

import scipy.stats as stats
```

```python
'''
O que √© Machine Learning?
Antes de falar de regress√£o linear, precisamos entender machine learning
(aprendizado de m√°quina).

Imagine que voc√™ quer ensinar um computador a prever algo, como o pre√ßo de uma
casa com base no tamanho dela. Em vez de dar uma f√≥rmula pronta, voc√™ mostra
v√°rios exemplos reais (dados) e o computador aprende sozinho uma forma de fazer
essa previs√£o.


O que √© Regress√£o Linear?
A regress√£o linear √© um dos algoritmos mais simples e mais importantes do
machine learning. Ela serve para prever um n√∫mero (um valor cont√≠nuo) com base
em um ou mais dados de entrada.
Exemplo:
Voc√™ tem um monte de informa√ß√µes como estas:
'''


'''
A√≠ voc√™ pergunta:

"Se eu tiver uma casa com 75m¬≤, qual ser√° o pre√ßo dela?"

A regress√£o linear tenta descobrir uma linha reta que melhor se ajusta aos dados
e permite fazer essa previs√£o.

Como funciona a linha?
A linha da regress√£o linear tem uma f√≥rmula parecida com a da matem√°tica:

y = a * x + b

x = valor de entrada (ex: tamanho da casa)
y = valor que queremos prever (ex: pre√ßo da casa)
a = inclina√ß√£o da linha (mostra o quanto o pre√ßo aumenta para cada m¬≤ a mais)
b = onde a linha cruza o eixo y (pre√ßo base quando o tamanho √© zero)

O algoritmo de regress√£o linear vai encontrar os valores de "a" e "b" que melhor
fazem essa linha passar perto dos pontos dos seus dados.

Como o computador aprende isso?

Voc√™ mostra os dados reais para o computador
(ex: v√°rias casas com tamanhos e pre√ßos).
Come√ßa com um chute qualquer para "a" e "b".
Compara os valores previstos com os valores reais e v√™ o erro.
Ajusta "a" e "b" um pouquinho para tentar diminuir esse erro.
Repete isso v√°rias vezes at√© encontrar a melhor linha.

Esse processo se chama treinamento.

Visualmente, como fica?

Se voc√™ colocar os dados em um gr√°fico:
No eixo X, o tamanho da casa
No eixo Y, o pre√ßo

Cada casa vira um ponto no gr√°fico.
A regress√£o linear desenha uma linha que tenta passar o mais pr√≥ximo
poss√≠vel desses pontos.

'''
```

```python
tamanho = [50, 60, 70, 80]

preco = 100000, 120000, 140000, 160000

tamanho = np.array(tamanho).reshape(-1, 1)

'''
np.array(tamanho) transforma a lista em array NumPy, que √© o formato que os
modelos aceitam.
.reshape(-1, 1) transforma o array em um formato de matriz com uma coluna,
porque o modelo espera esse formato: uma matriz de entrada. O par√¢metro (-1, 1)
serve para dizer ao numpy: Descubra sozinho quantas linhas eu preciso,
desde que tenha 1 coluna
'''


modelo  = LinearRegression()
modelo.fit(tamanho, preco)

'''
LinearRegression(): cria um modelo de regress√£o linear (reta).
.fit(tamanho, preco): o modelo aprende a rela√ß√£o entre o tamanho da casa e o
pre√ßo. Ou seja, ele ajusta uma linha que tenta passar o mais perto poss√≠vel dos
pontos dados.
'''

tamanho_para_plotar = np.linspace(tamanho.min(), tamanho.max(), 100).reshape(-1, 1)
preco_previsto = modelo.predict(tamanho_para_plotar)


'''
np.linspace(...): cria 100 valores entre o menor e o maior tamanho da casa.
Isso serve para desenhar a linha da regress√£o com suavidade no gr√°fico.

modelo.predict(...): o modelo prev√™ os pre√ßos para esses 100 tamanhos ‚Äî ou seja,
gera os pontos da linha vermelha do gr√°fico.
'''

# Plotando
plt.figure(figsize=(10, 6))
plt.scatter(tamanho, preco, color='blue', label='Casas (dados reais)')
plt.plot(tamanho_para_plotar, preco_previsto, color='red', linewidth=2, label='Linha de regress√£o')
plt.xlabel('Tamanho da casa (m¬≤)')
plt.ylabel('Pre√ßo da casa (R$)')
plt.title('Regress√£o Linear: Pre√ßo vs. Tamanho da Casa')
plt.legend()
plt.grid(True)
plt.show()
```

# REGRESS√ÇO MULTIPLA

```python
'''
Quando estamos tentando estimar y a partir de mais de uma vari√°vel independente,
estamos lidando com o que chamamos de Regress√£o Linear M√∫ltipla

Vamos entender passo a passo:
Na regress√£o linear simples, temos uma vari√°vel independente
(ex: o tamanho da casa).

F√≥rmula da regress√£o linear simples:
y = a * x + b

Mas na regress√£o linear m√∫ltipla, temos duas ou mais vari√°veis independentes
(ex: tamanho da casa e n√∫mero de quartos e localiza√ß√£o).

Exemplo pr√°tico:
Vamos imaginar uma planilha com dados sobre casas:
'''

'''
Agora queremos prever o pre√ßo da casa (y) com base em:
* Tamanho da casa
* N√∫mero de quartos
* Dist√¢ncia do centro da cidade

Como fica a f√≥rmula?
A f√≥rmula da regress√£o linear m√∫ltipla √© parecida, mas com mais vari√°veis:
y = a1 * x1 + a2 * x2 + a3 * x3 + ... + an * xn + b

Onde:
* x1, x2, ..., xn s√£o as vari√°veis independentes (entrada)
* a1, a2, ..., an s√£o os pesos ou coeficientes aprendidos pelo modelo
* b √© o intercepto (quanto vale y quando todas as vari√°veis forem zero)

No nosso exemplo:

Pre√ßo = a1 * Tamanho + a2 * Quartos + a3 * Dist√¢ncia + b

Como o computador aprende os valores de a1, a2, a3 e b?
Voc√™ fornece os dados (com entradas e sa√≠das).

O algoritmo come√ßa com valores aleat√≥rios para os coeficientes.
Calcula os pre√ßos previstos com esses valores.
Compara os valores previstos com os reais
(usa uma medida de erro, como o erro quadr√°tico m√©dio).
Ajusta os coeficientes para diminuir o erro.
Repete at√© encontrar os melhores valores poss√≠veis.

Esse processo tamb√©m √© chamado de treinamento.

Visualmente falando...
Com uma vari√°vel: a rela√ß√£o √© uma linha (2D).

Com duas vari√°veis: a rela√ß√£o vira um plano (3D).

Com mais de duas: vira um hiperplano
(dif√≠cil de imaginar, mas o conceito √© o mesmo).

Exemplo simples em linguagem natural:
Se o pre√ßo da casa depende de tamanho, n√∫mero de quartos e dist√¢ncia, ent√£o o
modelo vai aprender quanto cada fator influencia no pre√ßo.

Por exemplo:
Y = a1 * x1 + a2 * x2 +  a3 + x3 + b
Pre√ßo = 1500 * Tamanho + 20000 * Quartos - 10000 * Dist√¢ncia + 50000

Cada metro quadrado adiciona R$1.500
Cada quarto adiciona R$20.000
Cada km de dist√¢ncia tira R$10.000
E o valor base (intercepto) √© R$50.000

'''
```

### Cuidados essenciais ao Regres√£o Linear

```python
'''
1. Rela√ß√£o linear entre as vari√°veis
A regress√£o linear s√≥ funciona bem quando h√° uma rela√ß√£o linear entre as
vari√°veis.

‚úÖ O que √© isso?
Se voc√™ fizer um gr√°fico da vari√°vel x com a y, os pontos devem formar mais ou
menos uma linha reta (no caso simples), ou um padr√£o que possa ser ajustado por
um plano (no caso m√∫ltiplo).

‚ùå Se a rela√ß√£o for curva (ex: parab√≥lica, exponencial), a regress√£o linear vai
errar muito.
üëâ Dica: Use gr√°ficos de dispers√£o ou matriz de correla√ß√£o para verificar se a
rela√ß√£o √© linear.
'''
```

```python
'''
2. Aus√™ncia de multicolinearidade (no caso da m√∫ltipla)
Multicolinearidade acontece quando duas ou mais vari√°veis independentes s√£o
muito parecidas entre si.

‚ùå Problema:
O modelo "se confunde" e n√£o sabe qual vari√°vel est√° de fato influenciando o
resultado.

üõ† Como detectar:
Use a correla√ß√£o entre vari√°veis ou m√©tricas como o VIF
(Variance Inflation Factor).

'''
```

```python
'''
3. Homoscedasticidade (vari√¢ncia constante dos erros)
Os erros (diferen√ßa entre valor previsto e real) devem ter distribui√ß√£o parecida
ao longo de todos os valores de x.

‚ùå Se os erros aumentam ou diminuem com x, o modelo est√° com problema
(heteroscedasticidade).
üõ† Dica:
Fa√ßa um gr√°fico de res√≠duos (erros) vs. valores previstos. Eles devem parecer
aleat√≥rios, sem padr√£o.

'''
```

```python
'''
4. Erros devem ser normalmente distribu√≠dos
Isso √© importante principalmente quando usamos regress√£o para infer√™ncia
estat√≠stica (como p-valores ou intervalos de confian√ßa).

üõ† Dica:
Fa√ßa um histograma ou gr√°fico Q-Q dos res√≠duos (erros). Eles devem parecer uma
distribui√ß√£o normal (curva de sino).
'''
```

```python
'''
5. Aus√™ncia de outliers extremos
Outliers s√£o valores muito diferentes dos outros, e podem puxar a linha da
regress√£o para o lado errado.

üõ† Dica:
Use gr√°ficos de dispers√£o ou boxplots para detectar outliers. Se forem erros ou
exce√ß√µes, voc√™ pode:
* Remover
* Corrigir
* Transformar os dados

'''
```

```python
'''
6. Vari√°veis na mesma escala
Se voc√™ tiver vari√°veis com escalas muito diferentes
(ex: tamanho em m¬≤ e pre√ßo em milh√µes), isso pode atrapalhar o aprendizado do
modelo.

üõ† Dica:
Use padroniza√ß√£o ou normaliza√ß√£o (com StandardScaler, por exemplo).
'''
```

```python
'''
7. Quantidade de dados suficiente
Se voc√™ tiver poucos dados, o modelo pode se ajustar demais (overfitting) ou
n√£o aprender nada.

üõ† Dica:
Tente ter uma base de dados com amostras representativas e diversas.
'''
```

```python
'''
8. Separar dados de treino e teste
Nunca treine e teste o modelo com os mesmos dados.

‚úÖ Correto:
Separe os dados em conjunto de treino e teste
(por exemplo, 80% para treino, 20% para teste).

Assim voc√™ testa o desempenho real do modelo em dados novos.

'''
```

```python
'''
10. N√£o usar vari√°veis irrelevantes
Vari√°veis que n√£o t√™m rela√ß√£o com y s√≥ atrapalham o modelo.

üõ† Dica:
Use an√°lise de correla√ß√£o, sele√ß√£o de atributos (feature selection) ou
algoritmos de import√¢ncia de vari√°veis para identificar o que deve ou n√£o
entrar no modelo.
'''
```

```python
# Exemplo
'''
Objetivo geral do c√≥digo:
Queremos usar regress√£o linear m√∫ltipla para prever o pre√ßo de uma casa com base em:

Tamanho da casa (tamanho_m2)

N√∫mero de quartos (quartos)

Dist√¢ncia at√© o centro (distancia_centro)

E ao mesmo tempo, verificar se o modelo √© confi√°vel, passando por 10 testes importantes.
'''

# 0. Importando bibliotecas
'''
üìå Explica√ß√£o:

* pandas e numpy: para criar e manipular tabelas e n√∫meros.
* matplotlib.pyplot e seaborn: para criar gr√°ficos bonitos e informativos.
* scikit-learn (sklearn): para criar o modelo de regress√£o e avaliar a qualidade
  dele.
* statsmodels e scipy: para fazer testes estat√≠sticos mais avan√ßados
  (ex: normalidade dos erros, multicolinearidade).
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats
```

```python
# 1.1 Gerando dados artificiais com rela√ß√£o linear
'''
Explica√ß√£o:

* Cria uma base fict√≠cia de 200 casas com valores realistas.
* np.random.normal() gera n√∫meros parecidos com uma m√©dia
  (ex: tamanho m√©dio = 70 m¬≤).
* Calcula o pre√ßo com base na f√≥rmula da regress√£o linear.
* Adiciona um pouco de "erro" (+ np.random.normal(...)) para simular erros reais
'''

np.random.seed(42)
df = pd.DataFrame({
    'tamanho_m2': np.random.normal(70, 10, 200),
    'quartos': np.random.randint(1, 5, 200),
    'distancia_centro': np.random.normal(5, 2, 200)
})
df['preco'] = 50000 + 1500 * df['tamanho_m2'] + 20000 * df['quartos'] - 10000 * df['distancia_centro'] + np.random.normal(0, 10000, 200)
```

```python
df.head()
```

```python
# Verificando multicolinearidade (correla√ß√£o + VIF)
# 2.1 Matriz de correla√ß√£o
'''
üìå Explica√ß√£o:
Este gr√°fico mostra a correla√ß√£o entre as vari√°veis.
CorrelacÃßaÃÉo alta entre vari√°veis independentes (ex: tamanho e quartos) pode
indicar problemas de multicolinearidade.
'''
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correla√ß√£o")
plt.show()


# 2.2 Calculando VIF (Variance Inflation Factor)
'''
üìå Explica√ß√£o:

O VIF (Variance Inflation Factor) mede quanto uma vari√°vel est√° relacionada com
as outras.
Valores de VIF:
Abaixo de 5: tranquilo
Acima de 5: poss√≠vel alerta
Acima de 10: grave
'''
X_vif = df[['tamanho_m2', 'quartos', 'distancia_centro']]
X_vif = sm.add_constant(X_vif)

vif = pd.DataFrame()
vif["Vari√°vel"] = X_vif.columns
vif["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print("VIF (quanto maior, mais colinearidade):\n", vif)
```

```python
# Verificando homoscedasticidade dos erros
# 3.1 Separando X e y
'''
Explica√ß√£o:
X = entradas (vari√°veis independentes), y = sa√≠da (pre√ßo).
'''
X = df[['tamanho_m2', 'quartos', 'distancia_centro']]
y = df['preco']

# 3.2 Padronizando vari√°veis
'''
StandardScaler padroniza as vari√°veis para que fiquem na mesma escala.
'''
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3.3 Separando em treino e teste
'''
Explica√ß√£o:
Separa os dados:
80% para treinar o modelo
20% para testar depois
'''
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 3.4 Treinando o modelo
'''
Explica√ß√£o:
Cria e treina o modelo de regress√£o.
'''
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# 3.5 Previs√£o
'''
y_pred s√£o os valores previstos.
residuos √© o erro: diferen√ßa entre valor real e o previsto.
'''
y_pred = modelo.predict(X_test)
residuos = y_test - y_pred

# 3.6 Plotando res√≠duos vs valores previstos
'''
Explica√ß√£o:
Este gr√°fico mostra se os erros s√£o distribu√≠dos de forma uniforme.
Se formarem um "padr√£o", algo est√° errado. Se parecerem espalhados
aleatoriamente, est√° tudo bem!
'''
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_pred, y=residuos)
plt.axhline(0, color='red', linestyle='--')
plt.title("Res√≠duos vs Valores Previstos (deve parecer aleat√≥rio)")
plt.xlabel("Valores previstos")
plt.ylabel("Res√≠duos")
plt.show()
```

```python
# Verificar se os erros s√£o normais (forma de sino)


plt.hist(residuos, bins=20)
plt.title("Distribui√ß√£o dos Res√≠duos")
plt.show()

sm.qqplot(residuos, line='s')
plt.title("Q-Q Plot dos Res√≠duos")
plt.show()
```

```python
# # Verificar se os erros s√£o normais (forma de sino)
# Histograma
'''
Explica√ß√£o:
O histograma e o gr√°fico Q-Q servem para verificar se os erros seguem uma
distribui√ß√£o normal (curva de sino).
Isso √© importante em modelos estat√≠sticos!
'''
plt.hist(residuos, bins=20, edgecolor='k')
plt.title("Distribui√ß√£o dos Res√≠duos")
plt.xlabel("Erro")
plt.ylabel("Frequ√™ncia")
plt.show()

# Q-Q Plot
'''
Explica√ß√£o:

O histograma e o gr√°fico Q-Q servem para verificar se os erros seguem uma
distribui√ß√£o normal (curva de sino).
Isso √© importante em modelos estat√≠sticos!
'''
sm.qqplot(residuos, line='s')
plt.title("Q-Q Plot dos Res√≠duos")
plt.show()

# 4.3 Teste de Shapiro-Wilk
'''
Explica√ß√£o:
O teste de Shapiro-Wilk testa a normalidade dos erros.
Se o p-valor for maior que 0.05, os res√≠duos s√£o normalmente distribu√≠dos. ‚úÖ
'''
shapiro = stats.shapiro(residuos)
print(f"Shapiro-Wilk p-valor: {shapiro.pvalue:.4f} (ideal > 0.05 para normalidade)")
```

```python
#  Verificando outliers com boxplot
# Boxplots
'''
Explica√ß√£o:
Os boxplots mostram valores fora do padr√£o (outliers).
Um ou outro outlier √© normal, mas muitos podem prejudicar o modelo.
'''
for coluna in X.columns:
    sns.boxplot(data=df, x=coluna)
    plt.title(f"Boxplot de {coluna} (verificar outliers)")
    plt.show()
```

```python
# Verificar se temos dados suficientes
'''
Explica√ß√£o:
Em geral, √© bom ter pelo menos 10x mais linhas do que vari√°veis
(ex: 3 vari√°veis ‚Üí m√≠nimo 50 registros). Temos 200, ent√£o est√° √≥timo.
'''
print("Quantidade de registros:", len(df))
```

```python
# Avaliar o modelo com m√©tricas adequadas
'''
Explica√ß√£o:
MAE: Erro m√©dio absoluto Quanto menor melhor
MSE: Erro m√©dio quadrado (penaliza mais os erros grandes) Quanto menor melhor
RMSE: Raiz do erro quadrado (em R$) Quanto menor melhor
R¬≤: Mostra o quanto o modelo explica dos dados (1 = perfeito)
Escala do R¬≤:
1.00 = modelo perfeito (acertou tudo)
0.00 = modelo in√∫til (n√£o explica nada)
< 0.00 = pior que a m√©dia dos dados (modelo ruim mesmo!)
'''
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
media_preco = df['preco'].mean()
print(f"M√©dia dos pre√ßos das casas: R$ {media_preco:.2f}")
print("\nM√©tricas do modelo:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R¬≤: {r2:.4f}")
```
